```python
# Reshape B:
# Note If one the dimensions is set to -1, its size can be inferred
print(b.view(-1, 1))
print(b.view(12))
print(b.view(-1, 4))
print(b.view(3, 4))

# Assign b a new shape:
b = b.view(1, -1)
print(b)
print(b.shape)

# We can even reshape 3D tensors:
print('\n')

# create a 3D tensors:
three_dim = torch.randn(2, 3, 4)
print('\n')
print(three_dim)
print(three_dim.view(2, 12)) # reshape to 2 rows, 12 cols
print(three_dim.view(2, -1))
```

# Tensor Concatenation

```python
first_1 = torch.randn(2, 5)
print(first_1)
second_1 = torch.randn(3, 5)
print(second_1)

# Concatenate along dimension 0 (rows):
con_1 = torch.cat([first_1, second_1]) # This works correctly as both tensors have the same number of columns
print('\n')
print(con_1)
print('\n')

first_2 = torch.randn(2, 3)
print(first_2)
second_2 = torch.randn(2, 5)
print(second_2)

# Concatenate along dimension 1 (columns):
con_2 = torch.cat([first_2, second_2], 1) 1
print(con_2)
```
